```python
def load_data(df, table_name, engine):
    # Get the existing columns in the target table
    target_columns = pd.read_sql(f"SELECT * FROM {table_name} LIMIT 0", engine).columns
    
    # Select only the columns that exist in the target table
    df = df[target_columns.intersection(df.columns)]
    
    # Insert data into the target table
    with engine.connect() as connection:
        df.to_sql(table_name, connection, if_exists='append', index=False, method='multi')
```
